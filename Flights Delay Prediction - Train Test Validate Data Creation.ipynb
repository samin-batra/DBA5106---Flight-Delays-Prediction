```python
df_unique_airports = pd.DataFrame(df[['LATITUDE','LONGITUDE']].value_counts()).reset_index()
```

```python
df_delay_imbalance = pd.DataFrame(columns = ['LAT','LNG','DELAYED','NOT-DELAYED'])
print(df_delay_imbalance)
for index,row in df_unique_airports[['LATITUDE','LONGITUDE']].iterrows():
  delayed = df.loc[(df['LATITUDE']==row['LATITUDE']) & (df['LONGITUDE']==row['LONGITUDE'])]['DEP_DEL15'].value_counts().iloc[1]
  not_delayed = df.loc[(df['LATITUDE']==row['LATITUDE']) & (df['LONGITUDE']==row['LONGITUDE'])]['DEP_DEL15'].value_counts().iloc[0]
  # df_delay_imbalance.append({'LAT': row['LATITUDE'],'LNG':row['LONGITUDE'],'DELAYED':delayed, 'NOT-DELAYED':not_delayed},ignore_index=True)
  # pd.DataFrame(columns = ['LAT','LNG','DELAYED','NOT-DELAYED'])
  df_delay_imbalance = pd.concat([df_delay_imbalance,pd.DataFrame([{'LAT':row['LATITUDE'],'LNG':row['LONGITUDE'],'DELAYED':delayed,'NOT-DELAYED':not_delayed}])],ignore_index=True)

print(df_delay_imbalance)
df['DEP_DEL15'].value_counts()
```

```python
df_delay_imbalance['RATIO'] = df_delay_imbalance['DELAYED']/df_delay_imbalance['NOT-DELAYED']
print(df_delay_imbalance.sort_values(by='RATIO'))
lat = df_delay_imbalance.iloc[26]['LAT']
lng = df_delay_imbalance.loc[26]['LNG']
```

```python
df_training_data = df.loc[(df['LATITUDE']==lat) & (df['LONGITUDE']==lng)]
```

### Creating the data splits

```python
y = df_training_data['DEP_DEL15']
X = df_training_data.drop('DEP_DEL15',axis = 1)
from sklearn.model_selection import train_test_split

X_train, X_validate, y_train, y_validate = train_test_split(X, y, test_size=0.30, random_state=2022, stratify = y)

df_validate = pd.concat([X_validate, y_validate], axis = 1)
df_test = pd.concat([X_train, y_train],axis = 1)
```
