```python
X = df[['MONTH',
'DAY_OF_WEEK',
'DISTANCE_GROUP',
'DEP_TIME_BLK',
'SEGMENT_NUMBER',
'CONCURRENT_FLIGHTS',
'CARRIER_NAME',
'AIRLINE_AIRPORT_FLIGHTS_MONTH',
'AVG_MONTHLY_PASS_AIRPORT',
'AVG_MONTHLY_PASS_AIRLINE',
'FLT_ATTENDANTS_PER_PASS',
'GROUND_SERV_PER_PASS',
'PLANE_AGE',
'LATITUDE',
'LONGITUDE',
'PRCP',
'SNOW',
'SNWD',
'TMAX',
'AWND']]
```

```python
from sklearn.preprocessing import OneHotEncoder # Perform OneHotEnconding
```

```python
cat_vars_final = X.select_dtypes(['object','category'])
cat_vars_final
```

```python
enc = OneHotEncoder().fit(cat_vars_final)

cat_vars_ohe_final = enc.transform(cat_vars_final).toarray()
cat_vars_ohe_final = pd.DataFrame(cat_vars_ohe_final, index= cat_vars_final.index, 
                      columns=enc.get_feature_names(cat_vars_final.columns.tolist()))
```

```python
cat_vars_ohe_final
```

```python
Xnumeric = X.loc[:,~X.columns.isin(['DEP_TIME_BLK','CARRIER_NAME'])]
```

```python
df3 = cat_vars_ohe_final.join(Xnumeric)
```

```python
df3.shape
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators = 100, # number of trees
    max_depth= 5,       # parameters for the individual trees
    oob_score = True,   # whether or not to compute OOB score
    random_state = 22   # random state for reproducibility
)

rf.fit(df3,y)
```

```python
rf.feature_importances_
```

```python
pd.Series(rf.feature_importances_, index = rf.feature_names_in_).sort_values(ascending = False)
```

```python
rf.oob_score_
```

##### Looks great but the data was imbalanced!

```python
df['DEP_DEL15'].value_counts()
```

```python
# Class probablities for every sample
rf.oob_decision_function_
```

```python
# Threshold for predicting delay
t = 0.5

# OOB predictions for each sample
oob_preds = rf.oob_decision_function_[:,1] > t
oob_preds
```

```python
from sklearn.metrics import classification_report

print(classification_report(y, oob_preds)) # We are not testing on training set, so the results are accurate.
```

From this, we can see that the model isn't able to predict any Delays at all!

The recall of STAY is 0, and the balanced accuracy (macro avg recall) is 0.5!

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y, oob_preds, labels = [True, False])
```

Basically, we are predicting that ALL ARE DELAY?

```python
# Threshold for predicting DELAY
t = 0.1664  # <--- different threshold!

# OOB predictions for each sample
oob_preds = rf.oob_decision_function_[:,1] > t
oob_preds
```

```python
print(classification_report(y, oob_preds))
```

With just a simple adjustment, we have improved the balanced accuracy! We didn't even need to re-train the model!

Using the minority proportion is just a rule-of-thumb. It isn't guaranteed to give the best results.
